```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
```

```python
rcParams['figure.figsize'] = 8, 16
```

```python
reader = easyocr.Reader(['en', 'hi'])
```

```python
#file_name1 = "/content/drive/MyDrive/Major project/ezgif.com-gif-maker(2).jpg";
file_name2 = "/content/drive/MyDrive/Major project/ezgif.com-gif-maker(2).jpg";
```

```python
#Image(file_name1)
Image(file_name2)
```

```python
#output1 = reader.readtext(file_name1)
output2 = reader.readtext(file_name2)
```

```python
#output1
output2
```

```python
#cord1 = output1[1][0]
cord2 = output2[2][0]
```

```python
#x_min1, y_min1 = [int(min(idx)) for idx in zip(*cord1)]
x_min2, y_min2 = [int(min(idx)) for idx in zip(*cord2)]
```

```python
#x_max1, y_max1 = [int(max(idx)) for idx in zip(*cord1)]
x_max2, y_max2 = [int(max(idx)) for idx in zip(*cord2)]
```

```python
#image1 = cv2.imread(file_name1)
#cv2.rectangle(image1,(x_min1,y_min1),(x_max1,y_max1),(255,0,0),2)
#plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
image2 = cv2.imread(file_name2)
cv2.rectangle(image2,(x_min2,y_min2),(x_max2,y_max2),(255,0,0),2)
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
```
